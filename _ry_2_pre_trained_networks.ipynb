```python

from IPython.display import display
from PIL import Image

img1= Image.open('69724.jpg')
display(img1)
```

```python
img2= Image.open('69725.jpg')
display(img2)
```

```python
from torchvision import models
resnet= models.resnet101(pretrained=True)
resnet, \
resnet.eval()
```

```python

with open('imagenet_classes.txt') as f:
    labels= [line.strip() for line in f.readlines()]

len(labels), labels[0:10], labels[-10:]
```

```python
import torch
from torch import unsqueeze
from torchvision import transforms

preprocess= transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(
            mean=[0.485, 0.456, 0.406],
            std=[0.229, 0.224, 0.225]
        )])
```

```python
for x in   [img1, img2]:
    
    # show the input message
    
    print(f'x.filename= {x.filename}')
    display(x)
    
    # do the recognition 
    # from input  x 
    # to   output y
    
    y=  resnet(
            unsqueeze(
                preprocess(x), 
                0))
    
    # post processing 
    # to show the recognition results
    
    
    _, indices= torch.sort(y, descending=True)
    prob=       torch.nn.functional.softmax(y, dim=1)[0]
    theTop5=    [(labels[i], prob[i].item()) for i in indices[0][:5]]
    
    # show the recognition result
    
    #print(f'theTop5 candidates= {theTop5}')
    print(f'theTop5 candidates with probability:')
    for z in theTop5:
        print(f'{z}')
    print('--'*10)
```
